```python

# Merge the datasets on movie_id
merged_data = pd.merge(rating, items, on='movie_id')

# Calculate the average rating for each movie
average_ratings = merged_data.groupby('movie_id').agg({'rating': 'mean'}).reset_index()
average_ratings = average_ratings.rename(columns={'rating': 'average_rating'})

# Merge the average ratings with the item dataset
movie_data = pd.merge(items, average_ratings, on='movie_id')

# Function to get top 5 movies of a category
def get_top_5_movies(category):
    category_movies = movie_data[movie_data[category] == 1]
    top_5_movies = category_movies.sort_values(by='average_rating', ascending=False).head(5)
    return top_5_movies[['movie_id', 'movie title', 'average_rating']]

categories = ['Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

top_5_movies_per_category = {category: get_top_5_movies(category) for category in categories}

# Print the results
for category, top_movies in top_5_movies_per_category.items():
    print(f"Top 5 movies in {category}:")
    print(top_movies)
    print("\n")
```
